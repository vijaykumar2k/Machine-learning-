```python
# 2). DataFrame :----> It is multi-dimensional array and it returns values with column name
```

```python
#  Task ::: We have create a dataframe of employees in which column are
# emp_id, emp_name, department, salary, working_hours
```

```python
a = {
    "Emp_ID"        : ['A101','A102','A103','A104','A105','A106'],
    "Emp_Name"      : ['Sam','Rahul','Ankit','Gourav','Mohit','Karan'],
    "Department"    : ['HR','IT','HR','IT','HR','IT'],
    "Salary"        : [10000,20000,60000,55000,45000,78000],
    "Working_Hours" : [6,8,10,7,8,9]
}
```

```python
df = pd.DataFrame(a)
```

```python
df
```

```python
#  How can we export our dataframe into CSV data?
```

```python
df.to_csv("emp_info.csv")       ### to_csv means export your data into csv file
```

```python
df.to_csv("new_emp_info.csv", index=False)    ### index=False mean it will remove unnamed column.
```
